```python
+# importing useful libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read our dataset
df = pd.read_excel('/content/222.xlsx')
```

```python
df.head(5)
```

```python
df["Sconto"].fillna(df['Sconto'].mean(), inplace = True)
```

```python
df.head(5)
```

```python
# number of the each type of the gender
print(df.Facoltà.value_counts())
df.Facoltà.value_counts().plot.barh()
```

```python
lab = df['Residenza'].value_counts()
labels = lab.index
plt.figure(figsize=(6,10))
plt.pie(df['Residenza'].value_counts(),labels= labels, autopct='%.2f%%', startangle=90)
plt.title('participants by residente', fontweight='bold', fontsize=14)
plt.legend()
plt.legend(loc="upper left")
plt.show()
```
